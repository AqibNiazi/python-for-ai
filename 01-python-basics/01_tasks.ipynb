### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Step 1: Take inputs
heads = int(input("Enter total number of heads: "))
legs = int(input("Enter total number of legs: "))

# Step 2: Use equations
# Hint: From C + D = heads, we get C = heads - D

# Substitute into 2C + 4D = legs to solve for D

# Rearranging: 2(heads - D) + 4D = legs
# => 2*heads - 2*D + 4*D = legs
# => 2*heads + 2*D = legs
# => D = (legs - 2*heads) / 2
dogs = (legs - 2 * heads) // 2
chickens = heads - dogs

if dogs < 0 or chickens < 0 or (2*chickens + 4*dogs != legs):
    print("No valid solution")
else:
    print("Number of chickens:", chickens)
    print("Number of dogs:", dogs)
````
