```python
# Guessing game

# generate a random integer between 1 and 100
import random
jackpot = random.randint(1,100)

guess = int(input('guess karo: '))
counter = 1
while guess != jackpot:
  if guess < jackpot:
    print('Current guess {} !guess higher'.format(guess))
  else:
    print('Current guess {} !guess lower'.format(guess))

  guess = int(input('guess karo: '))
  counter += 1

else:
  print('correct guess')
  print('attempts',counter)


```

### Sequence sum

1/1! + 2/2! + 3/3! + ...

```python
import math

def sequence_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i / math.factorial(i)
    return total

# Example usage
n = 5
print("Sum of sequence up to", n, "terms is:", sequence_sum(n))
```

```python
# For loop vs While loops (When to use what?)
```

## For Loop
**Definition:** A for loop is used when you know exactly how many times you want to repeat something (iterate over a sequence like a list, string, or range).

**Best Use Cases:**

1. Iterating over a list, tuple, or dictionary

2. Running a loop for a fixed number of times

3. Cleaner and more readable when working with sequences

```python
for i in range(5):
    print(i)   # prints 0 to 4
```

## While Loop
**Definition:** A while loop is used when you don’t know in advance how many times you’ll need to repeat something. It runs until a condition becomes false.

**Best Use Cases:**

1. Waiting for user input until it matches a condition

2. Running until a certain state changes

3. Useful for indefinite repetition

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# Pattern 2
n = 4   # number of rows

for i in range(1, n + 1):
    # Increasing part
    for j in range(1, i + 1):
        print(j, end="")
    # Decreasing part
    for j in range(i - 1, 0, -1):
        print(j, end="")
    print()  # move to next line
```
